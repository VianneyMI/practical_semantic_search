```python
def get_some_titles(df:pd.DataFrame, titles:list[str])->pd.DataFrame:
    return df[df['song'].isin(titles)]

def get_some_artists(df:pd.DataFrame, artists:list[str])->pd.DataFrame:
    return df[df['artist'].isin(artists)]

def get_specific_song(df:pd.DataFrame, doc:dict)->pd.DataFrame:
    return df[(df['song'] == doc["title"]) & (df['artist'] == doc["artist"])]

def get_specific_songs(df:pd.DataFrame, docs:list[dict])->pd.DataFrame:
    dfs = []
    for doc in docs:
        dfs.append(get_specific_song(df, doc))

    return pd.concat(dfs, ignore_index=True)
```
